1. Use the **sticher** class


```python
import cv2
import glob
import matplotlib.pyplot as plt
import math
```

```python
# Read Images

boatImages = glob.glob("boat/*")
boatImages.sort()

boats = []
for b in boatImages:
  i = cv2.imread(b)
  i = cv2.cvtColor(i, cv2.COLOR_BGR2RGB)
  boats.append(i)

num_images = len(boats)
```

```python
# Display Images
plt.figure(figsize=[30,10]) 
num_cols = 3
num_rows = math.ceil(num_images / num_cols)
for i in range(0, num_images):
  plt.subplot(num_rows, num_cols, i+1) 
  plt.axis('off')
  plt.imshow(boats[i])
```

```python
# Stitch Images
stitcher = cv2.Stitcher_create()
status, result = stitcher.stitch(boats)
if status == 0:
  plt.figure(figsize=[30,10]) 
  plt.imshow(result)
```
